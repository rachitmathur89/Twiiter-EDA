<h3> Conclusion :</h3>- <b> In the above code created a function to sort the tweets words by words in a list.</b>

```python
lines = [re.sub(r'[^A-Za-z0-9]+', '', x) for x in lines]

lines

lines2 = []

for word in lines:
    if word != '':
        lines2.append(word)

s_stemmer = SnowballStemmer(language='english')

stem = []
for word in lines2:
    stem.append(s_stemmer.stem(word))
    
stem
```

<h4> Conclusion :</h4>- <b> Created a list for the number of words in the list created above</b>

```python
nlp = en_core_web_sm.load()
stem2 = []

for word in stem:
    if word not in nlp.Defaults.stop_words:
        stem2.append(word)

stem2

df = pd.DataFrame(stem2)
df = df[0].value_counts()
```
